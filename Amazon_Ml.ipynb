```python
csv_file_path = '/content/train.csv'
output_directory = 'resized_images'
resize_dimensions = (256, 256)
os.makedirs(output_directory, exist_ok=True)
```

```python
for index, row in dataset.iterrows():
    image_url = row[0]
    try:
        response = requests.get(image_url)
        response.raise_for_status()
        img_array = np.asarray(bytearray(response.content), dtype=np.uint8)
        img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
        if img is None:
            print('Image not found or cannot be decoded: {image_url}')
            continue
        resized_img = cv2.resize(img, resize_dimensions)
        base_name = os.path.basename(image_url)
        output_path = os.path.join(output_directory, base_name)
        cv2.imwrite(output_path, resized_img)
        print('Resized and saved: {output_path}')
    except Exception as e:
        print('Error processing {image_url}: {e}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model = VGG16(weights='imagenet', include_top=False, pooling='avg') # Pre existing model from the tensorflow used for the image preprocessing
```

```python
def extract_features_from_url(img_url):
    try:
        response = requests.get(img_url)
        response.raise_for_status()
        img = image.load_img(BytesIO(response.content), target_size=(224, 224),)
        x = image.img_to_array(img)
        x = preprocess_input(x)
        x = tf.expand_dims(x, axis=0)
        features = model.predict(x)
    except Exception as e:
        print(f'Error processing {img_url}: {e}')
        return None
```

```python
csv_file_path = '/content/train.csv'
test_data = pd.read_csv("/content/train.csv")

def extract_features(img_url):
    features = extract_features_from_url(img_url)

features_list = []
for index, row in test_data.iterrows():
    img_url = row[0]
    features = extract_features(img_url)
    features_list.append(features)

features_array = np.array(features_list)
```

```python
prediction = model.predict(features_array)
```
