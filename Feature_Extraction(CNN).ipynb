```python
import requests
from io import BytesIO
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
import tensorflow as tf
import numpy as np

# Load the pre-trained model (VGG16 in this case)
base_model = VGG16(weights='imagenet', include_top=False, pooling='avg')

def extract_features_from_url(img_url):
    try:
        # Fetch the image from the URL
        response = requests.get(img_url)
        response.raise_for_status()  # Raise an error for bad responses

        # Load the image and preprocess it
        img = image.load_img(BytesIO(response.content), target_size=(256, 256))
        x = image.img_to_array(img)
        x = preprocess_input(x)

        # Expand dimensions to match the model input shape
        x = np.expand_dims(x, axis=0)

        # Extract features from the image
        features = base_model.predict(x)

        # Flatten the features into a 1D array
        flattened_features = features.flatten()

        return flattened_features

    except Exception as e:
        print(f"Error processing image from URL: {img_url}")
        print(f"Error details: {e}")
        return None

features = extract_features_from_url(img_url)
print(features)
```
